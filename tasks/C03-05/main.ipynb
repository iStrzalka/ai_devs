```python
import sys

sys.path.append("../..")

from common.aidevs_auth import get_token_and_task, send_answer
from common.openai_auth import send_chat_completion
```

```python
token, task = get_token_and_task('people')
```

```python
task
```

```python
import json
import requests
import pandas as pd

data = json.loads(requests.get(task['data']).text)
```

```python
df = pd.DataFrame(data)
df['fullname'] = df['imie'] + ' ' + df['nazwisko']
df
```

```python
res = send_chat_completion(
    model_version='gpt-4', 
    system_content='You are taking a question. Return the NAME without diminutives and SURNAME of person within it.',
    user_content=task['question'],
)
```

```python
res['choices'][0]['message']['content']
```

```python
df[df['imie'] == 'Krystyna']
```

```python
record = df[df['fullname'] == res['choices'][0]['message']['content']].iloc[0]
```

```python
s = ""
for r in record.to_dict():
    s += f"{r}: {record[r]}\n"
```

```python
res = send_chat_completion(
    model_version='gpt-4',
    system_content="Be very specific and ultra-concise, return as little information as possible. When asked a question, return an equivalent sentce instead of full sentence" + s,
    user_content=task['question'],
)
res
```

```python
send_answer(token, res['choices'][0]['message']['content'])
```
